### Size

```python
fund['Estimated Shares Outstanding'] = fund['Estimated Shares Outstanding'].abs()
fund['size'] = np.log(fund['Estimated Shares Outstanding']*fund['close'])
```

```python
print(fund.shape)
fund.head()
```

```python
# NaN amount estimated shares outstanding
print(sum(fund['size'].isnull()))
for i, boo in enumerate(list(fund['size'].isnull())):
    if boo == True:
        print(i)
```

```python
fund['size'].describe()
```

### Non-linear Size

```python
fund['cube_size'] = fund['size']**3
fund.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
x = np.array(fund['size']).reshape(-1,1)
y = np.array(fund['cube_size']).reshape(-1,1)
regressor = LinearRegression()  
regressor.fit(x,y)
#To retrieve the intercept:
print("Inter:",regressor.intercept_)
#For retrieving the slope:
print("Coef:", regressor.coef_)

y_pred = regressor.predict(x)
print(y_pred)
res = y-y_pred
print(res)
```

```python
# fund['non linear size'] = np.NaN
fund['non linear size'] = res
```

```python
fund.head()
```

```python
fund['non linear size'].describe()
```

### Book-to-price

```python
fund['book to price'] = ((fund['Total Assets']-fund['Total Liabilities'])/fund['Estimated Shares Outstanding'])/fund['close']
```

```python
fund['book to price'].describe()
```

```python
print(sum(fund['book to price'].isnull()))
```

### Earning Yields-CETOP

```python
fund['EY-CETOP'] = fund['Earnings Per Share']/fund['close']
print(sum(fund['EY-CETOP'].isnull()))
fund['EY-CETOP'].describe()
```

### Earning Yields-ETOP

```python
fund['EY-ETOP'] = fund['Net Income']/(fund['Estimated Shares Outstanding']*fund['close'])
print(sum(fund['EY-ETOP'].isnull()))
fund['EY-ETOP'].describe()
```

### Leverage-MLEV

```python
fund['Leverage-MLEV'] = (fund['Estimated Shares Outstanding']*fund['close'] + fund['Long-Term Debt'])/(fund['Estimated Shares Outstanding']*fund['close'] )
print(sum(fund['Leverage-MLEV'].isnull()))
fund['Leverage-MLEV'].describe()
```

### Leverage-DTOA

```python
fund['Leverage-DTOA'] = fund['Total Liabilities']/fund['Total Assets']
print(sum(fund['Leverage-DTOA'].isnull()))
fund['Leverage-DTOA'].describe()
```

### Leverage-BLEV

```python
fund['Leverage-BLEV'] = (fund['Total Equity'] + fund['Long-Term Debt'])/ fund['Total Equity']
print(sum(fund['Leverage-BLEV'].isnull()))
fund['Leverage-BLEV'].describe()
```
