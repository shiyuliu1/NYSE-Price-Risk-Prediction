```python
import pandas as pd
import numpy as np
```

```python
fund = pd.read_csv('data/fundamentals.csv')
prices = pd.read_csv('data/prices.csv')
```

```python
def date_transfer(day):
    return day[0:10]
prices['date'] = prices.apply(lambda x: date_transfer(x['date']), axis=1)
prices.to_csv("prices_clean.csv", index = False)
fund.to_csv("fundamentals_clean.csv", index = False)
```

```python
fund = pd.read_csv('data/fundamentals_clean.csv')
prices = pd.read_csv('data/prices_clean.csv')
```

```python
display(fund.head())
print(fund.shape)
```

```python
display(prices.head())
print(prices.shape)
```

```python
new_df = pd.merge(fund, prices,  how='left', left_on=['Ticker Symbol','Period Ending'], right_on = ['symbol','date'])
display(new_df.head(10))
```

```python
new_df_na = new_df[new_df['date'].isnull()]
display(new_df_na.head())
print(new_df_na.shape)
new_df_ = new_df[new_df['date'].notnull()]
```

```python
new_df_na['Period Ending'] = (pd.DatetimeIndex(new_df_na['Period Ending']) - pd.offsets.Day(1)).astype(str)
display(new_df_na.head(10))
```

```python
new_df_na = new_df_na.drop(['date', 'symbol', 'open', 'close', 'low', 'high', 'volume'], axis = 1)
```

```python
new_df_na = pd.merge(new_df_na, prices,  how='left', left_on=['Ticker Symbol','Period Ending'], right_on = ['symbol','date'])
display(new_df_na.head(10))
```

```python
new_df_na2 = new_df_na[new_df_na['date'].isnull()]
display(new_df_na2.head())
print(new_df_na2.shape)
new_df_na_ = new_df_na[new_df_na['date'].notnull()]
```

```python
new_df_na2['Period Ending'] = (pd.DatetimeIndex(new_df_na2['Period Ending']) - pd.offsets.Day(1)).astype(str)
display(new_df_na2.head(10))
```

```python
new_df_na2 = new_df_na2.drop(['date', 'symbol', 'open', 'close', 'low', 'high', 'volume'], axis = 1)
new_df_na2 = pd.merge(new_df_na2, prices,  how='left', left_on=['Ticker Symbol','Period Ending'], right_on = ['symbol','date'])
display(new_df_na2.head(10))
```

```python
df_fillna = pd.concat([new_df_, new_df_na_, new_df_na2])
df_fillna = df_fillna.sort_values(by="Unnamed: 0")
df_fillna = df_fillna.reset_index()
df_fillna.drop(['index'],axis=1,inplace=True)
display(df_fillna.head(10))
print(df_fillna.shape)
```
